```python
# Count the number of titles released each year
titles_by_year = netflix_data['release_year'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(12, 6))
titles_by_year.plot(kind='bar')
plt.title('Number of Netflix Titles by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Count the distribution of content types (TV Show vs Movie)
content_types = netflix_data['type'].value_counts()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(content_types, labels=content_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Netflix Content Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
